## Now we have seen how this goes so let us just define a function which will do the same computation for 2017,2018,2019 and 2020

```python
import pandas as pd
import numpy as np


def process_pov_data(filename):
    
    '''This function takes the raw poverty file extracted from census data into account and it returns the poverty 
    percentage in different zipcodes by year ranging from 2016 - 2020'''

    # Read CSV file into pandas DataFrame
    data = pd.read_csv(filename)
    
    # Select every 6th column starting from the 5th column
    cols_n =[5]
    i = 5
    while i <len(data.columns):
        i = i+6
        cols_n.append(i)
    cols_n.pop()
    cols_n.insert(0, 0)
    
    # Select relevant columns, drop one column, and rename columns
    data = data.iloc[:,cols_n]
    data = data.drop(data.columns[2], axis = 1)
    data = data.rename(columns={data.columns[1] : data.columns[1][:13]})
    data = data.rename(columns={data.columns[2] : data.columns[2][:12]})
    for i in range(2,len(data.columns)):
        data = data.rename(columns={data.columns[i] : data.columns[i][6:11]})
    data = data.rename(columns = {data.columns[0] : 'Label'})
    
    # Transpose DataFrame, select rows, and rename columns
    data = data.iloc[[0]].transpose().reset_index()
    data = data.iloc[2:]
    pov = data.rename(columns = {'index':'zip',0:'Under poverty'}).reset_index(drop = True)
    
    # Replace missing values and convert percentage to decimal
    pov['Under poverty'] = pov['Under poverty'].replace(['-'],np.NaN)
    pov['Under poverty'] = pov['Under poverty'].str.rstrip('%').astype('float') / 100.0
    
    # Add year column and return processed DataFrame
    pov['year'] = int(filename[:4])
    return pov
```

```python
pov_2017 = process_pov_data('2017Pov_raw.csv')
pov_2018 = process_pov_data('2018Pov_raw.csv')
pov_2019 = process_pov_data('2019Pov_raw.csv')
pov_2020 = process_pov_data('2020Pov_raw.csv')
```

## Similarly, Extract population from the data from the first row

```python
def get_population_data(filename):
    """ Extract the population from the census data by pandas and string manipulation"""
    pov_raw = pd.read_csv(filename)
    i = 13         
    cols_n =[13]
    while i < len(pov_raw.columns):
        i = i + 6
        cols_n.append(i)

    cols_n.insert(0, 0)
    cols_n.pop()
    pov_raw = pov_raw.iloc[:, cols_n]

    for i in range(1, len(pov_raw.columns)):
        pov_raw = pov_raw.rename(columns={pov_raw.columns[i]: pov_raw.columns[i][6:11]}) 
    
    pov_raw = pov_raw.iloc[[0]].transpose().reset_index()
    pov_raw = pov_raw.iloc[1:]
    pov_raw = pov_raw.rename(columns = {'index': 'zip', 0: 'population'}).reset_index(drop=True)
    pov_raw['year'] = filename[0:4]
    return pov_raw[['zip', 'population', 'year']]
```

```python
pop_2016 = get_population_data('2016Pov_raw.csv')
pop_2017 = get_population_data('2017Pov_raw.csv')
pop_2018 = get_population_data('2018Pov_raw.csv')
pop_2019 = get_population_data('2019Pov_raw.csv')
pop_2020 = get_population_data('2020Pov_raw.csv')
```

```python
population = pd.concat([pop_2016,pop_2017,pop_2018,pop_2019,pop_2020])
population = population.reset_index(drop = True)
```

```python
population_poverty = pd.merge(population, poverty, on=['zip', 'year'])
```
