```python
# As there are 36 types of different crimes, let us combine them by their similarity and make 9 different categories
L1 = ["Homicide - Criminal","Homicide - Justifiable","Homicide - Gross Negligence"]
L2 = ["Other Assaults","Aggravated Assault No Firearm","Aggravated Assault Firearm"]
L3 = ['Thefts',"Theft from Vehicle","Motor Vehicle Theft","Receiving Stolen Property","Recovered Stolen Motor Vehicle"]
L4 = ["Rape","Other Sex Offenses (Not Commercialized)","Prostitution and Commercialized Vice"]
L5 = ['Fraud',"Forgery and Counterfeiting","Embezzlement"]
L6 = ["Vandalism/Criminal Mischief","Disorderly Conduct","Public Drunkenness","Vagrancy/Loitering","DRIVING UNDER THE INFLUENCE"]
L7 = ['Narcotic / Drug Law Violations',"Weapon Violations","Liquor Law Violations","Gambling Violations"]
L8 = ['All Other Offenses',"Arson","Offenses Against Family and Children"]
L9 = ["Burglary Non-Residential","Burglary Residential","Robbery Firearm","Robbery No Firearm"]
```

```python
from collections import ChainMap
d = ChainMap(dict.fromkeys(L1, "Homicide"), 
             dict.fromkeys(L2, 'Assaults'),
             dict.fromkeys(L3, "Theft"),
             dict.fromkeys(L4, "Sex Related"),
             dict.fromkeys(L5, "Fraud"),
             dict.fromkeys(L6, "Misconduct"),
             dict.fromkeys(L7, "Violation of Laws"),
             dict.fromkeys(L8, "Other"),
             dict.fromkeys(L9,"Burglary/Robbery"))
```

```python
crime_cleaned['crime_category'] = crime_cleaned['crime_type'].map(d.get) ## mapping
```

```python
# There are some latitudes for florida where lat<30 so only keeping values where lat>30
crime_cleaned = crime_cleaned[crime_cleaned['lat'] > 30].reset_index(drop = True)
```

```python
crime_cleaned.tail()
```

**Converting lat long for 750k rows would take a lot of time, so I have aggregated the data by crime_location field choosing the first long lat combination (because sometimes there is slight difference between some lat long for the same location but the zip code is always the same)**

**Aggregating records by location reduces the data points to convert from 750k to 54k**

**Now I will have to only convert 54k lat, long pair to zipcodes**

```python
trim_zip = crime_cleaned[['crime_location','lat','lng']]
## aggregate location and getting first value of long and lat
trim_zip_agg = trim_zip.groupby('crime_location').agg({'lat':'first','lng':'first'}) 
```

```python
trim_zip_agg = trim_zip_agg.reset_index()
```

```python
trim_zip_agg.tail()
```

```python
#pip install googlemaps
import googlemaps
```

```python
def get_zipcode(df, geolocator, lat_field, lon_field):
    location = gmaps.reverse_geocode((df[lat_field], df[lon_field]))
    ## This will return the zipcode. String slicing in formatted address
    return location[0]['formatted_address'][-10:-5]  
gmaps = googlemaps.Client(key='****************')
df = pd.DataFrame(trim_zip_agg)
```

```python
zipcodes = df.apply(get_zipcode, axis=1, geolocator=gmaps, lat_field='lat', lon_field='lng')
```

```python
merged_df = pd.merge(trim_zip_agg, zip_data, on=['lat', 'lng'], how='left')
```

```python
merged_df = merged_df.drop_duplicates().reset_index(drop = True)
```

```python
merged_df.tail()
```

```python
# As there were some non numeric and garbage values in zipcode column so convert everything into numeric
# Nin numberic values are specified as NaN
merged_df['google_zip'] = pd.to_numeric(merged_df['google_zip'], errors='coerce')
filtered_df = merged_df[(merged_df['google_zip'] >= 19100) & (merged_df['google_zip'] <= 19200)]
```

```python
filtered_df = filtered_df.drop_duplicates().reset_index(drop = True)
```

```python
all_crime_merged = crime_cleaned.merge(filtered_df[['crime_location','google_zip']], on = 'crime_location', 
                                       how = 'left') 
```

```python
all_crime_merged = all_crime_merged.drop_duplicates().reset_index(drop = True)
```
